```python
from rdflib import Namespace, Graph, URIRef, Literal, XSD
from rdflib.namespace import OWL
```

```python
file_path = "/home/ubuntu/yago-data/"
```

```python
yago_file_list = ['yago-wd-full-types.nt','yago-wd-shapes.nt','yago-wd-labels.nt','yago-wd-simple-types.nt','yago-wd-class.nt','yago-wd-sameAs.nt','yago-wd-facts.nt','yago-wd-schema.nt']
```

```python
%time
for i in yago_file_list:
    G = Graph()
    G.parse(file_path + i, format="nt")
    G.serialize(destination=f'/home/ubuntu/vol1/yago-turtles/{i[:-3]}.ttl')
    print(i)
```

```python
i = 1
for s, p, o in G.triples((None, None, None)):
    print(s,p,o)
    i+=1
    if i >= 1000:
        break
```

```python
query = "SELECT ?s ?p ?o WHERE{?s a <http://schema.org/ChildCare>. ?s ?p ?o} LIMIT 50"
result = G.query(query)
for row in result:
    print(row,"\n")
```
